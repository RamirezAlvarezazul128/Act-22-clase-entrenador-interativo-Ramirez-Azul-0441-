## Clase entrenador interactivo

```python
class Entrenador:
    def __init__(self, nombre, deporte, años_experiencia):
        self.nombre = nombre
        self.deporte = deporte
        self.años_experiencia = años_experiencia

    def calentar(self):
        print(f"\n[Calentamiento] {self.nombre} ordena: '¡Todos a calentar para la sesión de {self.deporte}!'")

    def explicar_ejercicio(self):
        print(f"[Técnica] {self.nombre} está demostrando el ejercicio principal.")

    def motivar(self):
        print(f"[Motivación] {self.nombre} dice: '¡No se rindan! Llevo {self.años_experiencia} años en esto y sé que pueden.'")

# Función para solicitar datos y crear un objeto Entrenador
def crear_entrenador(numero):
    print(f"\n--- Ingrese los datos para el Entrenador {numero} ---")
    nombre = input("Nombre: ")
    deporte = input("Deporte: ")
    experiencia = input("Años de experiencia: ")

    # Retornamos el objeto creado con los datos ingresados
    return Entrenador(nombre, deporte, experiencia)

# --- Flujo Principal ---

# Pedimos datos para el primer entrenador
entrenador1 = crear_entrenador(1)

# Pedimos datos para el segundo entrenador
entrenador2 = crear_entrenador(2)

# --- Ejecución de acciones ---

print("\n" + "="*30)
print("DEMOSTRACIÓN DE FUNCIONES")
print("="*30)

# Acciones del Entrenador 1
entrenador1.calentar()
entrenador1.explicar_ejercicio()
entrenador1.motivar()

# Acciones del Entrenador 2
entrenador2.calentar()
entrenador2.explicar_ejercicio()
entrenador2.motivar()

print("fin del programa")
print("programa realizado por Ramirez Azul NL = 44")
```
